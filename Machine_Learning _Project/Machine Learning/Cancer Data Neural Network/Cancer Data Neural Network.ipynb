```python

# encode the categorial varibles to numerical by using LabelEncoder:

from sklearn.preprocessing import LabelEncoder
lr= LabelEncoder()


# col=['gender','smoking_history']
# for i in col:
df['diagnosis']=lr.fit_transform(df['diagnosis']) # fit_transform is using to calculate the mapping between the unique categories in the data and their corresponding numerical labels and apply it to tranform the input data.
df=df.drop('Unnamed: 32',axis=1)
```

```python
df.head(5)
df.tail(10)
```

```python

X=df.drop('diagnosis',axis=1)
Y=df['diagnosis']
```

```python
# Standardizing the data before using machine learning algorithm. Tranforming the variables to the same scale:

from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_s=ss.fit_transform(X)
```

```python
# Split the feature set (X_s) and the target set (Y) into traning and testing set.
# xtr is the training feature set, xte is the testing feature set, 
# ytr is the training target set, yte is the testing target set.
# test size 0.8 means 80% data is using for testing and 20% data is using for training

from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(X_s,Y,test_size=0.8) 
```

```python
from sklearn.tree import DecisionTreeClassifier

# creating an instance of `DecisionTreeClassifier`
clf_dt=DecisionTreeClassifier()

# Train the DecisionTreeClassifier using training data.
clf_dt.fit(xtr,ytr)

# Predicting the target labels for the test data:
ypre_dt=clf_dt.predict(xte)
```

```python
# from sklearn.neural_network import
#alpha: prevent ovefitting by adding a penalty term to the loss function. default value is 0.0001
# 5 hidden layers, and 2 neurons.
# random state: number seed of reproducibility
from sklearn.neural_network import MLPClassifier

clf_nn = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)

# Training the MLP Classifier using the training data 
# Using the fit method of the MLPClassifier instance (clf_nn) to train the neural network using the training features (xtr) and the corresponding target labels (ytr).
clf_nn.fit(xtr,ytr)

# Predict the target labels for the test data. using the trained MLPClassifier to make predictions on the test features (xte), and the predicted target labels are stored in ypre_nn.
ypre_nn=clf_nn.predict(xte)
```

```python
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix for the decision tree classifier (ypre_dt):
cm1=confusion_matrix(yte,ypre_dt)

# Compute the confusion matrix for the multi-layer perceptron classifier (ypre_nn)
cm2=confusion_matrix(yte,ypre_nn)
```

```python
from sklearn.metrics import accuracy_score

# Compute and print the accuracy score for the decision tree classifier
print(accuracy_score(yte,ypre_dt))

# Compute and print the accuracy score for the multi-layer perceptron classifier
print(accuracy_score(yte,ypre_nn))
```

- The accuracy of approximately of 0.92 means that the decision tree classifier correctly classified 92.11% of samples in the test set.

- The accuracy of approximately of 0.96 means that the multi-layer perceptron classifier correctly classified 95.83% of samples in the test set.

In comparision, multi-layer perceptron classifier (0.9583) has higher accuracy score than decision tree classifier(0.92105)

The higher accuracy score of multi-layer perceptron classifier suggests that it performed slightly better in terms of predictions on the test data compared to the decision tree classifier.
